```scala
import sml.{Agent, Identifier, WMElement}
    
    
  trait WmeEncoder[A] {
    def encode(agent: Agent, parent: Identifier, attr: String, value: A): Unit
  }
    
  trait WmeDecoder[A] {
    def decode(id: Identifier, attr: String): A
  }
                                                                                                
    
  implicit val stringEncoder: WmeEncoder[String] = new WmeEncoder[String] {
    def encode(agent: Agent, parent: Identifier, attr: String, value: String): Unit =
      agent.CreateStringWME(parent, attr, value)
  }
    
  implicit val longEncoder: WmeEncoder[Long] = new WmeEncoder[Long] {
    def encode(agent: Agent, parent: Identifier, attr: String, value: Long): Unit =
      agent.CreateIntWME(parent, attr, value)
  }
implicit val doubleEncoder: WmeEncoder[Double] = new WmeEncoder[Double] {                                                                                                                                                              
    def encode(agent: Agent, parent: Identifier, attr: String, value: Double): Unit =                                                                                                                                                    
      agent.CreateFloatWME(parent, attr, value)                                                                                                                                                                                          
  }                                                                                                                                                                                                                                      
                                                                                                                                                                                                                                         
  implicit val stringDecoder: WmeDecoder[String] = new WmeDecoder[String] {                                                                                                                                                              
    def decode(id: Identifier, attr: String): String =                                                                                                                                                                                   
      id.GetParameterValue(attr)                                                                                                                                                                                                         
  }                                                                                                                                                                                                                                      
                                                                                                                                                                                                                                         
  implicit val longDecoder: WmeDecoder[Long] = new WmeDecoder[Long] {                                                                                                                                                                    
    def decode(id: Identifier, attr: String): Long =                                                                                                                                                                                     
      id.FindByAttribute(attr, 0).ConvertToIntElement().GetValue()                                                                                                                                                                       
  }                                                                                                                                                                                                                                      
                                                                                                                                                                                                                                         
  implicit val doubleDecoder: WmeDecoder[Double] = new WmeDecoder[Double] {                                                                                                                                                              
    def decode(id: Identifier, attr: String): Double =                                                                                                                                                                                   
      id.FindByAttribute(attr, 0).ConvertToFloatElement().GetValue()                                                                                                                                                                     
  }                          
```

```scala
def wmeWrite[A](agent: Agent, parent: Identifier, attr: String, value: A)                                                                                                                                                              
                 (implicit enc: WmeEncoder[A]): Unit = enc.encode(agent, parent, attr, value)                                                                                                                                                                                               
                                                                                                                                                                                                                                         
def wmeRead[A](id: Identifier, attr: String)                                                                                                                                                                                           
                (implicit dec: WmeDecoder[A]): A =  dec.decode(id, attr)                                                                                                                                                                                                                 
                                                                                                                                                                                                                                                                                                                                                                                                                                                                             
```

```scala
case class Location(x: Double, y: Double, label: String)                                                                                                                                                                               
                                                                                                                                                                                                                                         
object Location {                                                                                                                                                                                                                   
    def writeTo(agent: Agent, parent: Identifier, loc: Location): Identifier = { 
      val id = agent.CreateIdWME(parent, "location")                                                                                                                                                                                     
      wmeWrite(agent, id, "x",     loc.x)                                                                                                                                                                                                
      wmeWrite(agent, id, "y",     loc.y)                                                                                                                                                                                                
      wmeWrite(agent, id, "label", loc.label)                                                                                                                                                                                            
      id                                                                                                                                                                                                                     
    }                                                                                                                                                                                                                                    
                                                                                                                                                                                                                                         
    def readFrom(id: Identifier): Location = Location(
      x     = wmeRead[Double](id, "x"),                                                                                                                                                                                                  
      y     = wmeRead[Double](id, "y"),                                                                                                                                                                                                  
      label = wmeRead[String](id, "label")                                                                                                                                                                                               
    )                                                                                                                                                                                                                                  
  }  
```
